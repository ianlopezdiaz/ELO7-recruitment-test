````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
from unidecode import unidecode

def clean_text(text):
    text = str(text)
    text = unidecode(text)
    text = text.lower()
    for t in ',.:;-_/\\+=()[]<>^~?!#*%':
        text = text.replace(t, ' ')
    return ' '.join(text.split())

def clean_and_separate_tags(text):
    text = clean_text(text)
    tags = [t for t in text.split() if not t.isdigit()]
    return tags

# carregando os dados
raw_data = pd.read_csv("elo7_recruitment_dataset.csv")
df = raw_data.copy()

# lidando com dados faltantes
df['order_counts'].fillna(0, inplace=True)
df['concatenated_tags'].fillna('', inplace=True)
df.dropna(inplace=True)
df.drop(['product_id','seller_id'], axis=1).describe()

# definindo novas variaveis
df['creation_date'] = pd.to_datetime(df['creation_date'])
date_today = pd.to_datetime('today')
df['number_of_days'] = [(date_today-day).days for day in df['creation_date']]

df['len_tags'] = [len(clean_text(t)) for t in df['concatenated_tags']]
df['len_title'] = [len(clean_text(t)) for t in df['title']]
df['len_query'] = [len(clean_text(t)) for t in df['query']]
df['price_per_weight'] = df['price']/(df['weight']+1)

df['order_probability'] = df['order_counts']/df['view_counts']
df = df[df['order_probability'] <= 1]

df['views_per_day'] = df['view_counts']/df['number_of_days']
df['orders_per_day'] = df['order_counts']/df['number_of_days']

df.drop(['product_id','seller_id'], axis=1).describe()
````

Feita essa preparação inicial dos dados, segue a lista das variáveis (features) disponíveis para a análise.

````python
qualitative_columns = ['seller_id',
                       'product_id',
                       'creation_date',
                       'query',
                       'title',
                       'concatenated_tags',
                       'category',
                       'express_delivery']

quantitative_columns_original = list(raw_data.drop([*qualitative_columns], axis=1).columns)
quantitative_columns_new = list(df.drop([*(qualitative_columns+quantitative_columns_original)], axis=1).columns)
quantitative_columns = quantitative_columns_original + quantitative_columns_new

print('Variáveis qualitativas:')
for t in qualitative_columns:
    print(f'    - {t}')
print('\nVariáveis quantitativas do dataset original:')
for t in quantitative_columns_original:
    print(f'    - {t}')
print('\nVariáveis quantitativas criadas:')
for t in quantitative_columns_new:
    print(f'    - {t}')
````

E agora? Como eu vou saber se o o modelo treinado está classificando decentemente?

Há algumas métricas que são comumente usadas para avaliar o desempenho dos modelos. As mais comuns são: **accuracy (acurácia)**, **precision (precisão)**, **recall** e **F1 score**. Estas são todas calculadas comparando-se a previsão do modelo usando os dados de teste (`y_pred`) com o valor já previamente conhecido para esta variável de saída (`y_test`). Assim os resultados do modelo são classificados em *positivos verdadeiros (true positives)* (TP), *positivos falsos (false positives)* (FP), *negativos verdadeiros (true negatives)* (TN) e *negativos falsos (false negatives)* (FN). As métricas acima são calculadas emfunção dos números correspondentes a cada caso (TP, FP, TN, FN) da seguinte forma:

- $\mbox{accuracy} = \frac{TP+TN}{TP+FP+FN+TN}$
- $\mbox{pecision} = \frac{TP}{TP+FP}$
- $\mbox{recall}   = \frac{TP}{TP+FN}$
- $\mbox{F1 score} = 2\times\frac{\mbox{precision}\times\mbox{recall}}{\mbox{precision}+\mbox{recall}}$

É importante notar que as definições destas métricas de avaliação são referentes à classificação binária, ou seja, para os casos em que estes modelos são usados para prever se algo **está** ou **não está** em uma dada categoria. Quando temos um sistema de classificação em mais de uma categoria estas métricas são calculadas por categoria e podem ser calculadas para o modelo como um todo através de alguma média.

Segue então o desempenho desta versão inicial do classificador random forest:

````python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred0))
````

Uma acurácia de 63%. Tem espaço para melhorar bastante. Claro, isso é só uma rodada de familiarização com os modelos.

Ainda há algo interessante que pode ser feito nesta etapa. Eu posso usar um método para verificar a importância de cada feature para esse modelo. Este método (**permutation_importance**) repete o treino e o teste do modelo com diferentes subconjuntos das features e reavalia as métricas. Segue o resultado:

````python
from sklearn.inspection import permutation_importance
result0 = permutation_importance(rf0, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2)
sorted_idx = result0.importances_mean.argsort()

fig, ax = plt.subplots(figsize=(10, 3))
ax.boxplot(result0.importances[sorted_idx].T, vert=False, labels=X_test.columns[sorted_idx])
ax.set_title("Permutation Importances (RF - original features)")
fig.tight_layout()
plt.show()
````

Podemos ver claramente aqui que a variável `price` tem uma relevância muito maior na capacidade de previsão deste modelo, ao passo que `psosition`, por exemplo, tem importância quase nula.

## 2.3. Um básico de processamento de linguagem natural (NLP)

Tudo o que eu escrevi sobre como lidar com as variáveis textuais na [análise exploratória](01_analise_exploratoria.ipynb) foi feito antes que eu fosse pesquisar como de fato funciona o processamento de linguagem natural. Felizmente nada do que foi feito nesta primeira parte do trabalho será útil para esta nova etapa. Algumas das coisas que eu fiz neste começo, como converter os textos para minúsculas, remover acentos, remover pontuação, remover símbolos não alfanuméricos etc, até estavam no caminho certo, mas ainda assim há mais passos a serem levados em consideração. De qualquer forma, o texto da análise exploratória vai continuar como está, afinal, as tentativas que deram errado também devem ser documentadas.

As etapas a serem seguidas para preparar as variáveis textuais para a análise são:

1. homogeneização (converter todo o texto para minúsculas)
1. remover caracteteres irrelevantes (pontuação e símbolos não alfanuméricos)
1. remover palavras que não acrescentam informação ao texto, as chamadas stopwords ('de', 'a', 'o', 'e', 'que' etc)
1. converter palavras escritas incorretamente para um padrão (ex: oooooiiiiii - oi)
1. lematização (transformar palavras como 'sou', 'és', 'é', 'somos' em 'ser')
1. tokenização (separar o texto em palavras)

O pacote usado para fazer a maioria destas coisas é o [`nltk`](https://www.nltk.org/). Porém infelizmente este pacote não tem módlos para lematização e correção de palavras incorretas em português. Ele até tem uma lista de stopwords em português, o que será muito útil, mas os ítens 4 e 5 destas etapas de preparação do texto não serão feitas aqui.

Após esta preparação do texto, é necessário usar algum método para converter o texto em algum formato numérico para que ele possa ser usado como input do modelo que queremos treinar. Isto pode ser feito de várias formas diferentes, como os seguintes métodos de vetorização:

- Bag of Words: neste método, o texto é transformado em um vetor que nos dá informação sobre a multiplicidade de cada palavra no texto, mas não sobre a ordem de aparição das palavras ou estrutura semântica. Este modelo apenas conta quantas vezes cada palavra aparece no texto.
- TF-IDF: este método é um pouco mais sofisticado do que o Bag of Words. O nome do método vem de Term Frequency - Inverse Document Frequency, que são duas métricas calculadas para fornecer um valor numérico a cada palavra do documento. Estas duas métricas são calculadas como:
    - $\mbox{TF(t)} = \frac{\mbox{número de vezes que a palavra t aparece no documento}}{\mbox{número de palavras no documento}}$
    - $\mbox{IDF(t)} = \log_e{(\frac{\mbox{número total de documentos}}{\mbox{número de documentos com a palavra t}})}$ o valor de TF-IDF é o produto entre TF e IDF. Este valor é uma medida do quão importante cada palavra é dentro de uma coleção de textos. A importância da palavra então é proporcional à sua prequência em um determinado documento mas diminui se a palavra for muito frequente no conjunto total de documentos.
- Hashing: uma das limitações dos métodos anteriores é que o número de palavras em um conjunto de textos pode ser muito grande, desta forma armazenar as frquências de todas as palavras em um vetor pode ser muito custoso, isso sem contar que o esforço computacional para usar este vetor como input em um modelo de machine learning pode tornar este processo impraticável dependendo do quão grande for o volume de texto. Um jeito de contornar este problema é usar o "hashing-trick", no qual não se guarda o vocabulário do conjunto de textos e se atribui um número fixo arbitrário de "palavras". Estas "palavras" serão mapeadas em uma matriz com o tamanho correspondente ao número de "palavras" escolhido e este mapeamento é feito usando uma função hash ([Murmurhash3](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.FeatureHasher.html)). A vantagem de se usar este método é a economia de memória e tempo computacional. A desvantagem é que não há como fazer a transformação inversa e recuperar o texto original a partir da matriz gerada neste processo, coisa que pode ser feita nos outros métodos.


Pela discussão acima sobre os métodos de vetorização de texto já dá para ter uma noção de que as tentativas com o TF-IDF não deram muito certo. Então vamos aos detalhes. Primeiro eu vou 


````python
import nltk
from nltk.corpus import stopwords
#nltk.download('stopwords') ## apenas no primeiro uso!!!!!!
from nltk.tokenize import word_tokenize
text_query = [' '.join([tt for tt in word_tokenize(t)
                        if tt not in stopwords.words('portuguese')]) for t in df['query']]
````

A lista `text_query` agora contém todos os valores de `query` dos usuários. Os textos já estão homogeneizados, livres de pontuação, acentos e símbolos irrelevantes, bem como livres de stopwords. Logo, estão prontos para ser vetorizados. Vou começar com o método Hashing, gerando algumas matrizes de tamanhos 5, 10, 15, 20, 30 e 40 só para visualizar o resultado do método.

````python
from sklearn.feature_extraction import FeatureHasher

hasher = {}
hash_query_df = {}
for n_features in (5, 10, 15, 20, 25, 30, 40, 50, 60, 100, 200):
    hasher[n_features] = FeatureHasher(n_features=n_features, input_type='string')
    query = hasher[n_features].fit_transform(text_query)
    hash_query_df[n_features] = pd.DataFrame(query.toarray(), columns=[f'q{i:d}' for i in range(n_features)])
    print(hash_query_df[n_features])
````

## 2.4. Implementação e avaliação dos modelos: agora com a feature textual `query` e a feature numérica `price`

Agora que a variável `query` está vetorizada vetorizada, falta implementar os modelos com esse novo dataset.
Para ter uma noção da diferença o número de features de `hash_query_df` faz para os modelos, vou fazer um teste apenas usando apenas o classificador random forest para os vários números de features e comparar as métricas de avaliação de cada um.

````python
y = df['category']
for n_features in hash_query_df:
    X = hash_query_df[n_features]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    clf_rf  = RandomForestClassifier(n_estimators=10)
    clf_rf.fit(X_train, y_train)
    y_pred = clf_rf.predict(X_test)
    print()
    print('-'*70)
    print(f'n_features = {n_features}')
    print(classification_report(y_test, y_pred, zero_division=1))
````

A princípio a acurácia do modelo aumenta conforme o número de features, porém a partir de 30 features a acurácia parece atingir um platô. Realmente 5 ou 10 features parece pouco, mas usar 200 features não melhora a qualidade do modelo (pelo menos pelas métricas usadas) e aumenta consideravelmente o tempo de execução. Desta forma um valor razoável parece ser 30 features.

Agora é possível usar o dataframe `hash_query_df[30]` como input para treinar e testar todos os modelos de novo. Dessa forma posso comparar o desempenho dos diversos modelos com o mesmo conjunto de dados. Além disso, também vou testar a diferença que faz no desempenho de cada modelo se os dados forem padronizados (usando `StandardScaler`) ou normalizados (usando o `MinMaxScaler`). É importante ressaltar que o modelo SGD sequer rodou com os dados sem padronizar ou normalizar.

Seguem as comparações entre as métricas dos modelos Random Forest (RF), k-Nearest Neighbors (KNN), Suport Vector Machines (SVM) e Stochastic Gradient Descent (SDG).

````python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

n_features = 30

X = hash_query_df[n_features]
y = df['category']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


clf = {}

clf['RF (no scaling)']  = RandomForestClassifier(n_estimators=10)
clf['RF (StandardScaler)']  = make_pipeline(StandardScaler(), RandomForestClassifier(n_estimators=10))
clf['RF (MinMaxScaler)']  = make_pipeline(MinMaxScaler(), RandomForestClassifier(n_estimators=10))

clf['KNN (no scaling)'] = KNeighborsClassifier(n_neighbors=7)
clf['KNN (StandardScaler)'] = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7))
clf['KNN (MinMaxScaler)'] = make_pipeline(MinMaxScaler(), KNeighborsClassifier(n_neighbors=7))

clf['SVM (no scaling)'] = svm.SVC(kernel='rbf', C=10)
clf['SVM (StandardScaler)'] = make_pipeline(StandardScaler(), svm.SVC(kernel='rbf', C=10))
clf['SVM (MinMaxScaler)'] = make_pipeline(MinMaxScaler(), svm.SVC(kernel='rbf', C=10))

#clf['SGD (no scaling)'] = make_pipeline(StandardScaler(), SGDClassifier(max_iter=1000, tol=1e-3))
clf['SGD (StandardScaler)'] = make_pipeline(StandardScaler(), SGDClassifier(max_iter=1000, tol=1e-3))
clf['SGD (MinMaxScaler)'] = make_pipeline(MinMaxScaler(), SGDClassifier(max_iter=1000, tol=1e-3))



# trinar, testar e imprimir as métricas
y_pred = {}
for k in clf:
    clf[k].fit(X_train, y_train)
    y_pred[k] = clf[k].predict(X_test)
    print()
    print('-'*60)
    print(f'{k:20s}', '-'*39)
    print(classification_report(y_test, y_pred[k], zero_division=1))
````

Usando a acurácia como parâmetro de comparação, é possível dizer que o RF é levemente melhor do que seus pares independente se os dados estiverem normalizados, padronizados ou sem qualquer transformação. O SGD performou pior do que todos os outros. O SVM performou um pouco melhor do que o RF apenas quando seus dados foram padronizados, mas a diferença é mínima.

Ainda é possível melhorar o desempenho do modelo SVM. Basta usar um método para procurar os valores dos parâmetros ótimos do modelo para este dataset. O código para fazer esta tarefa usando o `GridSearchCV` é 

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
grid = GridSearchCV(svm.SVC(), param_grid, refit = True, verbose = 3)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.best_estimator_)
```
onde pode-se acrecentar mais valores para serem testados nas listas tando do parâmetro `C` quanto do parâmetro `gamma`. O problema deste método é que ele é bastante demorado e no fim das contas a acurácia do modelo não melhora muito (aumenta de 82% para 83% usando esses valores). Mesmo que aumentasse para 85% ou mais próximo de 90%, ainda assim, se esse modelo tivesse de ser retreinado no futuro quando mais dados estivessem disponíveis, esse procedimento para encontrar os melhores parâmetros deveria ser executado novamente e dependendo da frequência com que esse modelo tivesse de ser retreinado isso acabaria se tornando muito custoso. Nessas condições, o Random Forest simplesmente se sai muito bem com muito menos trabalho para tunar parâmetros e sequer é necessário padronizar ou normalizar os dados.

Há ainda outra forma interessante de comparar o desempenho destes modelos. Isto pode ser feito de uma forma tanto numérica quanto visual através das curvas ROC (Receiver Operating Characteristic Curve). Esta curva representa o desempenho de um sistema classificador binário. A curva é um gráfico da taxa de verdadeiros positivos (FPR) contra a taxa de falsos positivos (TPR) do classificador.

Seguem os gráficos de curvas ROC para os modelos. Estes gáficos foram feitos usando o pacote [yellowbrick](https://www.scikit-yb.org/en/latest/).

````python
from yellowbrick.classifier import ROCAUC

fig, axs = plt.subplots(4, 3, figsize=(20, 20))

labels = [
        ['RF (no scaling)', 'RF (StandardScaler)', 'RF (MinMaxScaler)'],
        ['KNN (no scaling)', 'KNN (StandardScaler)', 'KNN (MinMaxScaler)'],
        ['SVM (no scaling)', 'SVM (StandardScaler)', 'SVM (MinMaxScaler)'],
        [None, 'SGD (StandardScaler)', 'SGD (MinMaxScaler)']
]

ROC_visualizer = {}
for i in range(4):
    for j in range(3):
        k = labels[i][j]
        if k:
            ROC_visualizer[k] = ROCAUC(clf[k], classes=clf[k].classes_, ax=axs[i][j])
            ROC_visualizer[k].fit(X_train, y_train)
            ROC_visualizer[k].score(X_test, y_test)
            axs[i][j].legend(loc='lower right')
            axs[i][j].set_title(k)
            axs[i][j].set_xlabel('FPR')
            axs[i][j].set_ylabel('TPR')
        else:
            axs[i][j].axis('off')

fig.tight_layout(pad=3.0)
plt.show()
````

Basicamente, uma curva ROC ideal seria uma reta vertical em FPR=0 de TPR=0 a TPR=1 seguida de uma reta horizontal de FPR=0 a FPR=1. Assim a taxa de verdadeiros positivos seria 100% e a taxa de falsos positivos seria zero. A linha diagonal tracejada corresponde aos casos nos quais a taxa de verdadeiros positivos é igual à taxa de falsos positivos, ou seja, um classificador aleatório. Logo, quanto mais longe as curvas ROC estiverem desta diagonal melhor é o classificador. O caso extremo da curva ideal também teria uma integral igual a 1. O valor desta integral está nas legendas (AUC, area under curve).

Novamente, é  possível ver que os desempenhos dos três Random Forest, dos três k-Nearest Neighbors e do Suport Vector Machines com dados padronizados são equivalentes. Nos demais casos já fica evidente que a classificação de algumas categorias como "Papel e Cia" e "Outros" acaba sendo mais imprecisa. Aliás, a classificação destas duas categorias é mais imprecisa em todos os modelos mas é especialmente ruim no SGD.

## Conclusão

As duas impementações do modelo SGD foram reprovadas em todas as métricas. Pelas curvas ROC ainda podemos descartar o SVM com dados normalizados e o SVM sem transformações. Usando a acurácia como critério, os Random Forest performam melhor do que os k-Nearest Neighbors e empatam com o Suport Vector Machines com dados padronizados. Porém, como já foi discutido, gastando-se mais tempo seria possível encontrar parâmetros que melhorem a acurácia do SVM, porém a desvantagem é a necessidade de tunar estes parâmetros sempre que for necessário retreinar o modelo com dados novos. Sendo assim, a escolha com a melhor relação custo/benefício é um dos Random Forest, podendo inclusive ser a versão sem padronização ou normalização de dados, o que é ainda melhor para o tempo de execução.
